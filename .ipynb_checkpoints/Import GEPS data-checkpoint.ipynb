```python
def load_data(data_path):
    f = io.matlab.loadmat(data_path)
    loop_data = f['Loopdata_mixed'][()]
    X = loop_data.reshape(loop_data.shape[0]*loop_data.shape[0],-1)
    X -= np.nanmean(X)
    X[np.where(np.isfinite(x)==0)] = 0
    X /= np.std(X)
    X = np.atleast_3d(X)
    return X
```
